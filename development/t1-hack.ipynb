```python
def process_sprint_entities(sprint_df, entity_df):
    """
    Разгруппировка сущностей по спринтам
    
    Parameters:
    sprint_df (pandas.DataFrame): Датафрейм спринтов с колонками 'sprint_name' и 'entity_ids'
    entity_df (pandas.DataFrame): Датафрейм сущностей с колонкой 'entity_id'
    
    Returns:
    dict: Словарь, где ключи - названия спринтов, значения - датафреймы с сущностями для каждого спринта
    """
    # Проверка на наличие необходимых колонок
    if 'sprint_name' not in sprint_df.columns or 'entity_ids' not in sprint_df.columns:
        raise ValueError("sprint_df должен содержать колонки 'sprint_name' и 'entity_ids'")
    if 'entity_id' not in entity_df.columns:
        raise ValueError("entity_df должен содержать колонку 'entity_id'")

    # Создаем словарь для хранения результатов
    sprint_entities = {}
    
    # Проходим по каждой строке в таблице спринтов
    for _, row in sprint_df.iterrows():
        # Получаем название спринта
        sprint_name = row['sprint_name']
        
        # Получаем entity_ids как множество, если оно уже в виде set
        if isinstance(row['entity_ids'], set):
            entity_ids = row['entity_ids']
        else:
            # Если entity_ids в строковом формате, конвертируем в множество
            entities_str = row['entity_ids'].strip('{}')
            entity_ids = {int(id_str.strip()) for id_str in entities_str.split(',') if id_str.strip()}
        
        # Фильтруем датафрейм сущностей по полученным ID
        sprint_entities[sprint_name] = entity_df[entity_df['entity_id'].isin(entity_ids)].copy()
    
    return sprint_entities

# Получаем разгруппированные данные
result = process_sprint_entities(sprints, df)

```

```python
from ast import literal_eval


# Преобразуем строковые представления множеств в настоящие множества
def safe_literal_eval(val):
    if isinstance(val, set):
        return val
    return literal_eval(val)

sprints['entity_ids'] = sprints['entity_ids'].apply(safe_literal_eval)

# Разгруппируем entity_ids в sprints, чтобы получить по одному идентификатору на строку
sprints_exploded = sprints.explode('entity_ids')
```
