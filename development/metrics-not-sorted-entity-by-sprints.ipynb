```python
import pandas as pd
from datetime import datetime

sprints_info = pd.DataFrame(sprints)
sprints_info["sprint_start_date"] = pd.to_datetime(sprints_info["sprint_start_date"])
sprints_info["sprint_end_date"] = pd.to_datetime(sprints_info["sprint_end_date"])
tasks = pd.DataFrame(entities)
tasks["update_date"] = pd.to_datetime(tasks["update_date"])

# Функция для выделения спринтов и расчета метрик
def calculate_metrics(timestamp, sprints, tasks):
    """
    Рассчитывает метрики для активного и прошедших спринтов.
    
    Args:
        timestamp (str): Таймлайн (дата в формате 'YYYY-MM-DD').
        sprints (pd.DataFrame): DataFrame с колонками ["sprint_name", "start_date", "end_date"].
        tasks (pd.DataFrame): DataFrame с колонками ["task_id", "status", "update_date", "estimation"].
    
    Returns:
        dict: Словарь с метриками для активного и прошедших спринтов.
    """
    # Преобразуем timestamp в datetime
    current_time = pd.to_datetime(timestamp)
    
    # Определяем активный спринт
    active_sprint = sprints[(sprints_info["sprint_start_date"] <= current_time) & (sprints["sprint_end_date"] >= current_time)]
    
    # Определяем прошедшие спринты
    past_sprints = sprints[sprints_info["sprint_end_date"] < current_time]
    
    # Рассчитать метрику "К выполнению" для активного спринта
    if not active_sprint.empty:
        active_sprint = active_sprint.iloc[0]
        active_tasks = tasks[
            (tasks["update_date"] >= active_sprint["sprint_start_date"]) &
            (tasks["update_date"] <= active_sprint["sprint_end_date"]) &
            (tasks["status"] == "Создано")
        ]
        active_metric = active_tasks["estimation"].sum() / 3600  # Перевод в часы
    else:
        active_metric = None  # Нет активного спринта
    
    # Рассчитать метрику "К выполнению" для каждого прошедшего спринта
    past_metrics = []
    for _, sprint in past_sprints.iterrows():
        past_tasks = tasks[
            (tasks["update_date"] >= sprint["sprint_start_date"]) &
            (tasks["update_date"] <= sprint["sprint_end_date"]) &
            (tasks["status"] == "Создано")
        ]
        past_metric = past_tasks["estimation"].sum() / 3600  # Перевод в часы
        past_metrics.append({"sprint_name": sprint["sprint_name"], "metric": past_metric})
    
    # Формируем результат
    result = {
        "active_sprint": {
            "sprint_name": active_sprint["sprint_name"] if not active_sprint.empty else None,
            "metric": active_metric,
        },
        "past_sprints": past_metrics,
    }
    
    return result

# Пример вызова функции
timestamp = "2024-09-12"
result = calculate_metrics(timestamp, sprints_info, tasks)
print(result)
```

```python
import pandas as pd
from datetime import datetime

sprints_info = pd.DataFrame(sprints)
sprints_info["sprint_start_date"] = pd.to_datetime(sprints_info["sprint_start_date"])
sprints_info["sprint_end_date"] = pd.to_datetime(sprints_info["sprint_end_date"])
tasks = pd.DataFrame(entities)
tasks["update_date"] = pd.to_datetime(tasks["update_date"])
# Функция для расчёта показателя "Сделано"

def calculate_done_metric(timestamp, sprints, tasks):
    """
    Рассчитывает показатель "Сделано" для активного и прошедших спринтов.
    
    Args:
        timestamp (str): Таймлайн (дата в формате 'YYYY-MM-DD').
        sprints (pd.DataFrame): DataFrame с колонками ["sprint_name", "start_date", "end_date"].
        tasks (pd.DataFrame): DataFrame с колонками ["task_id", "status", "update_date", "estimation"].
    
    Returns:
        dict: Словарь с показателями для активного и прошедших спринтов.
    """
    # Преобразуем timestamp в datetime
    current_time = pd.to_datetime(timestamp)
    
    # Определяем активный спринт
    active_sprint = sprints[(sprints["sprint_start_date"] <= current_time) & (sprints["sprint_end_date"] >= current_time)]
    
    # Определяем прошедшие спринты
    past_sprints = sprints[sprints["sprint_end_date"] < current_time]
    
    # Рассчитать показатель "Сделано" для активного спринта
    if not active_sprint.empty:
        active_sprint = active_sprint.iloc[0]
        # Для активного спринта считаем только задачи, изменившие статус на "Закрыто" или "Выполнено"
        active_tasks = tasks[
            (tasks["update_date"] >= active_sprint["sprint_start_date"]) & 
            (tasks["update_date"] <= current_time) &  # Динамическое обновление до текущего времени
            (tasks["status"].isin(["Закрыто", "Выполнено"]))  # Статусы "Закрыто" или "Выполнено"
        ]
        active_metric = active_tasks["estimation"].sum() / 3600  # Перевод в часы
    else:
        active_metric = None  # Нет активного спринта
    
    # Рассчитать показатель "Сделано" для каждого прошедшего спринта
    past_metrics = []
    for _, sprint in past_sprints.iterrows():
        past_tasks = tasks[
            (tasks["update_date"] >= sprint["sprint_start_date"]) & 
            (tasks["update_date"] <= sprint["sprint_end_date"]) &  # Учитываем задачи в пределах спринта
            (tasks["status"].isin(["Закрыто", "Выполнено"]))  # Статусы "Закрыто" или "Выполнено"
        ]
        past_metric = past_tasks["estimation"].sum() / 3600  # Перевод в часы
        past_metrics.append({"sprint_name": sprint["sprint_name"], "metric": round(past_metric, 1)})
    
    # Формируем результат
    result = {
        "active_sprint": {
            "sprint_name": active_sprint["sprint_name"] if not active_sprint.empty else None,
            "metric": round(active_metric, 1) if active_metric is not None else None,
        },
        "past_sprints": past_metrics,
    }
    
    return result

# Пример вызова функции
timestamp = "2024-09-17"
result = calculate_done_metric(timestamp, sprints_info, tasks)
print(result)
```

```python
def calculate_removed_metric(timestamp, sprints, tasks):
    """
    Рассчитывает метрику "Снято" для активного или выбранного прошлого спринта.
    
    Args:
        timestamp (str): Таймлайн (дата в формате 'YYYY-MM-DD').
        sprints (pd.DataFrame): DataFrame с колонками ["sprint_name", "start_date", "end_date"].
        tasks (pd.DataFrame): DataFrame с колонками ["task_id", "status", "resolution", "update_date", "estimation", "is_defect"].
    
    Returns:
        dict: Метрика "Снято" для активного или выбранного прошлого спринта.
    """
    # Преобразуем timestamp в datetime
    current_time = pd.to_datetime(timestamp)
    
    # Определяем активный спринт
    active_sprint = sprints[(sprints["start_date"] <= current_time) & (sprints["end_date"] >= current_time)]
    
    # Если спринт активный
    if not active_sprint.empty:
        sprint = active_sprint.iloc[0]
        is_active = True
    else:
        # Определяем последний завершённый спринт
        past_sprints = sprints[sprints["end_date"] < current_time]
        if past_sprints.empty:
            return {"sprint_name": None, "removed_metric": 0.0}  # Нет спринтов для расчёта
        sprint = past_sprints.iloc[-1]
        is_active = False
    
    # Фильтрация задач, относящихся к выбранному спринту
    sprint_tasks = tasks[
        (tasks["update_date"] >= sprint["start_date"]) &
        (tasks["update_date"] <= sprint["end_date"])
    ]
    
    # Условия для "снятых" объектов
    if is_active:
        # Для активного спринта — учитывать ежедневные изменения статусов
        removed_tasks = sprint_tasks[
            ((sprint_tasks["status"].isin(["Закрыто", "Выполнено"])) &
             (sprint_tasks["resolution"].isin(["Отклонено", "Отменено инициатором", "Дубликат"]))) |
            ((sprint_tasks["is_defect"] == True) & (sprint_tasks["status"] == "Отклонен исполнителем"))
        ]
    else:
        # Для прошедшего спринта — учитывать последний статус в его временных рамках
        removed_tasks = sprint_tasks[
            ((sprint_tasks["status"].isin(["Закрыто", "Выполнено"])) &
             (sprint_tasks["resolution"].isin(["Отклонено", "Отменено инициатором", "Дубликат"]))) |
            ((sprint_tasks["is_defect"] == True) & (sprint_tasks["status"] == "Отклонен исполнителем"))
        ]
    
    # Расчёт метрики "Снято"
    removed_metric = removed_tasks["estimation"].sum() / 3600  # Перевод в часы с округлением до одного знака
    removed_metric = round(removed_metric, 1)
    
    return {
        "sprint_name": sprint["sprint_name"],
        "removed_metric": removed_metric
    }

# Пример данных задач
tasks_data = [
    {"task_id": 1, "status": "Закрыто", "resolution": "Отклонено", "update_date": "2024-07-12", "estimation": 7200, "is_defect": False},
    {"task_id": 2, "status": "Выполнено", "resolution": "Отменено инициатором", "update_date": "2024-07-12", "estimation": 3600, "is_defect": False},
    {"task_id": 3, "status": "Отклонен исполнителем", "resolution": None, "update_date": "2024-07-16", "estimation": 5400, "is_defect": True},
    {"task_id": 4, "status": "Закрыто", "resolution": "Дубликат", "update_date": "2024-07-20", "estimation": 1800, "is_defect": False},
]

tasks = pd.DataFrame(tasks_data)
tasks["update_date"] = pd.to_datetime(tasks["update_date"])

# Пример вызова функции
timestamp = "2024-07-17"
result = calculate_removed_metric(timestamp, sprints, tasks)
print(result)
```
